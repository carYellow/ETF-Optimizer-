## 1. Generate Technical Indicators

Technical indicators are mathematical calculations based on price, volume, or open interest information. They are widely used to identify trends, momentum, and potential reversal points in stock prices. Common examples include moving averages, RSI, MACD, and Bollinger Bands.

```python
from src.data.data_loader import StockDataLoader
# Initialize data loader and load a subset of S&P 500 stock data
data_loader = StockDataLoader(start_date="2010-01-01")
symbols = data_loader.get_sp500_symbols()
sample_symbols = symbols[:5]  # Use a small subset for demonstration
stock_data, sp500_data = data_loader.prepare_training_data(sample_symbols)

# Use stock_data for feature analysis
df = stock_data.copy()
# If Date is a column, set as index for time series analysis
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)

print(f"Loaded stock data shape: {df.shape}")
display(df.head())
```

```python
# The DataFrame 'df' is loaded using StockDataLoader in the previous cell.
# Ensure it contains columns: ['Open', 'High', 'Low', 'Close', 'Volume']

# Moving Averages
for window in [5, 20, 50]:
    df[f'SMA_{window}'] = df['Close'].rolling(window=window).mean()

# Relative Strength Index (RSI)
delta = df['Close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
rs = gain / loss
df['RSI_14'] = 100 - (100 / (1 + rs))

# MACD
ema12 = df['Close'].ewm(span=12, adjust=False).mean()
ema26 = df['Close'].ewm(span=26, adjust=False).mean()
df['MACD'] = ema12 - ema26
df['MACD_signal'] = df['MACD'].ewm(span=9, adjust=False).mean()

# Bollinger Bands
window = 20
df['BB_Middle'] = df['Close'].rolling(window=window).mean()
df['BB_Std'] = df['Close'].rolling(window=window).std()
df['BB_Upper'] = df['BB_Middle'] + 2 * df['BB_Std']
df['BB_Lower'] = df['BB_Middle'] - 2 * df['BB_Std']

# Plot Moving Averages and Bollinger Bands
plt.figure(figsize=(14, 6))
plt.plot(df['Close'], label='Close')
plt.plot(df['SMA_20'], label='SMA 20')
plt.plot(df['BB_Upper'], label='Bollinger Upper', linestyle='--', color='grey')
plt.plot(df['BB_Lower'], label='Bollinger Lower', linestyle='--', color='grey')
plt.title('Close Price, SMA, and Bollinger Bands')
plt.legend()
plt.show()

# Plot RSI
plt.figure(figsize=(14, 3))
plt.plot(df['RSI_14'], label='RSI 14', color='purple')
plt.axhline(70, color='red', linestyle='--')
plt.axhline(30, color='green', linestyle='--')
plt.title('Relative Strength Index (RSI)')
plt.legend()
plt.show()

# Plot MACD
plt.figure(figsize=(14, 3))
plt.plot(df['MACD'], label='MACD', color='blue')
plt.plot(df['MACD_signal'], label='Signal Line', color='orange')
plt.title('MACD and Signal Line')
plt.legend()
plt.show()
```

## 2. Create Price and Volume-Based Features

Price and volume-based features capture relationships between OHLCV (Open, High, Low, Close, Volume) values. These features can help identify unusual price movements, volatility, and liquidity changes. Common examples include price ratios, volatility measures, and volume spikes.

```python
# OHLC Ratios

df['High_Low_Range'] = df['High'] - df['Low']
df['Close_Open_Change'] = df['Close'] - df['Open']
df['HL_Pct'] = (df['High'] - df['Low']) / df['Low']
df['CO_Pct'] = (df['Close'] - df['Open']) / df['Open']

# Volatility (Rolling Standard Deviation of Returns)
df['Return'] = df['Close'].pct_change()
df['Volatility_10'] = df['Return'].rolling(window=10).std()
df['Volatility_20'] = df['Return'].rolling(window=20).std()

# Volume Spike (Z-score)
df['Volume_Z'] = (df['Volume'] - df['Volume'].rolling(window=20).mean()) / df['Volume'].rolling(window=20).std()

# Plot Volatility
plt.figure(figsize=(14, 4))
plt.plot(df['Volatility_10'], label='10-day Volatility')
plt.plot(df['Volatility_20'], label='20-day Volatility')
plt.title('Rolling Volatility of Returns')
plt.legend()
plt.show()

# Plot Volume Z-score
plt.figure(figsize=(14, 4))
plt.plot(df['Volume_Z'], label='Volume Z-score', color='brown')
plt.axhline(2, color='red', linestyle='--', label='Spike Threshold')
plt.title('Volume Spike Detection (Z-score)')
plt.legend()
plt.show()
```

## 3. Analyze Trend and Momentum Signals

Trend and momentum features help identify the direction and strength of price movements. These features are useful for detecting persistent trends or reversals. Examples include momentum, rate of change (ROC), and directional movement indicators.

```python
# Momentum (n-period difference)
df['Momentum_10'] = df['Close'] - df['Close'].shift(10)

# Rate of Change (ROC)
df['ROC_10'] = df['Close'].pct_change(periods=10)

# Directional Movement Index (DMI) approximation
# For full DMI/ADX, use a library like ta or talib. Here, we approximate directional movement.
df['UpMove'] = df['High'] - df['High'].shift(1)
df['DownMove'] = df['Low'].shift(1) - df['Low']
df['+DM'] = np.where((df['UpMove'] > df['DownMove']) & (df['UpMove'] > 0), df['UpMove'], 0)
df['-DM'] = np.where((df['DownMove'] > df['UpMove']) & (df['DownMove'] > 0), df['DownMove'], 0)

# Plot Momentum and ROC
plt.figure(figsize=(14, 4))
plt.plot(df['Momentum_10'], label='10-day Momentum', color='teal')
plt.title('Momentum (10-day)')
plt.legend()
plt.show()

plt.figure(figsize=(14, 4))
plt.plot(df['ROC_10'], label='10-day Rate of Change', color='navy')
plt.title('Rate of Change (10-day)')
plt.legend()
plt.show()
```
